### Use given Iris dataset and solve the following quesions:

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
iris.keys()
```

```python
df=pd.DataFrame(iris["data"])
df.columns=iris["feature_names"]
df["Target"]=iris["target"]
df
```

### Your goal is to predict the species in a given dataset by considering target variable as "target" 
and feature variable as "petal length".

```python
x=df[["petal length (cm)"]]
x
```

```python
y=df["Target"]
y
```

### 5. Apply Linear Regression to the given dataset without splitting dataset and Check the R squared score

```python
from sklearn.linear_model import LinearRegression
lineModel=LinearRegression()
```

```python
lineModel.fit(x,y)
```

```python
lineModel.score(x,y)
```

```python
from sklearn.metrics import r2_score
```

```python
y
```

```python
y_predict=lineModel.predict(x)
y_predict
```

```python
r2_score(y,y_predict)
```

### 6. Split the dataset into training and testing sets and Check the R squared score

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
lineModel=LinearRegression()
lineModel.fit(x_train,y_train)
print("training score")
print("\t",lineModel.score(x_train,y_train))
print("test score")
print("\t",lineModel.score(x_test,y_test))
```
